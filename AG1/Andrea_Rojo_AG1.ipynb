## Devolución de cambio por técnica voraz

```python
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor #valor_Acumulado es 0, primera iteracion del bucle
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

## N-Reinas por técnica de vueta atrás


```python
# restricciones: no podian compartir ni fila, ni columna ni diagonal entre reinas
#para comprobar de manera eficiente que no se violan las restricciones:
# - Comprobando diagonales calculando diferencia entre posiciones de índices y valores en valor absoluto
# - Reinas tienen que tener distinto indice de fila, se modela con un array, y esas posiciones son la altura de las reinas en le tablero.

def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  print('*****')
  print(SOLUCION)
  escribe(SOLUCION)
  print('*****')
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1: #la estructura del array soluciones (0, 2,2,2) Esto devolveria false a la funcion prometedora.
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False #comprobar todas las reinas que tenemos disponibles. #(0,2,1,3)//(0,1,2,3) Devolveria false porque hay reinas enfrentadas.
  return True
#hay que desarrollar un arbol, que es una estructura de datos, que almacene el camino que segui hasta que no se cumpla la condicion prometedora, hay que parar la expansion del arbol.


def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(4)
```

## Viaje por el rio. Programación dinámica

```python
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]

#cuando tiene infinito por que es un buen modelo para indicar que no hay conexion directa? Porque se tiene en cuenta que es un problema de minimizar los costes y la distancia.
#tarea revisr matriz, si partimos del modo 1 solo podemos llegar al 3,al 4 y al 6.

################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N] #el nodo es correlativo ala profundidad, es decir primero se llega al nodo 3 y luego al 4.
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j] #i=0, j=6
      RUTA[i][j] = i #i=0

      for k in range(i, j): #para optimizar el calculo se mete el segundo bucle.
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")#almacena cual es siguiente mejor salto para llegar hasta el destino.
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
#el resultado que aparece es el camino con menos peso
```

## Práctica Individual. Encontrar los dos puntos más cercanos

```python
import math
#Problema resuelto mediante el algoritmo voraz.
def dos_puntos_cercanos_voraz(puntos):
    minima_distancia=float('inf')
    puntos_dist_min=None
    for i in range(len(puntos)):
        for j in range(i+1, len(puntos)):
            distancia=math.dist(puntos[i], puntos[j])
            if distancia < minima_distancia:
                minima_distancia=distancia
                puntos_dist_min=(puntos[i], puntos[j])

    return minima_distancia, puntos_dist_min

puntos=[(0,0), (1,5), (2,4), (3,7)]
minima_distancia, puntos_dist_min=dos_puntos_cercanos_voraz(puntos)
print(f'La distancia mínima es {minima_distancia} y los puntos son {puntos_dist_min}')
```

En este caso, de fuerza bruta, la complejidad es O(n^2). Mediante el método de Divide y vencerás si que se puede mejorar la complejidad porque reduce el espacio de búsqueda, ya que no busca el punto cercano en todo el espacio sino donde es posible que exista.

```python
def merge_sort_1d(lista):
    result=[]
    #si la lista solo tiene un elemento devuelve la lista
    if len(lista)<=1:
        return lista
    #Separa la lista por la mitad, la parte de la izquierda y de la derecha.
    izq_lista=merge_sort_1d(lista[:(len(lista)//2)])
    der_lista=merge_sort_1d(lista[(len(lista)//2):])
    i=0
    j=0
    #se recorre toda la lista
    for _ in range(len(der_lista)+len(izq_lista)):
        #si ya se ha gastado toda la lista derecha, se añade lo que falte a la izquierda
        if j==len(der_lista):
            result.append(izq_lista[i])
            i=i+1
        #si ya se ha gastado toda la lista izquierda, se añade lo que falte a la derecha
        elif i==len(izq_lista):
            result.append(der_lista[j])
            j=j+1
        #si los dos índices son válidos, entonces se compara las dos lista y se añade al que menos elementos tiene.
        elif der_lista[j]<= izq_lista[i]:
            result.append(der_lista[j])
            j=j+1
        else:
            result.append(izq_lista[i])
            i=i+1

    return result
```

```python
from math import inf

def dos_puntos_cercanos_divide_venceras_1d(puntos):
    #Ordenar los puntos
    ordenados=merge_sort_1d(puntos)
    # Si el tamaño del conjunto es 2, se va a devolver la distancia entre ellos. Se devolverá infinito si tiene 1 o ningún elemento.
    if len(puntos)<2:
        return (None, inf)
    if len(puntos)==2:
        return ((puntos[0], puntos[1]), math.dist(puntos[0], puntos[1]))

    #Se inicializa la distancia minima como infinito y los puntos más cercanos a None.
    mejores_puntos= None
    dist_minima=inf

    #Recorrer la lista de puntos
    for i in range(len(puntos)-1):
        #se calcula la distancia entre los puntos adyacentes
        dist=ordenados[i+1]-ordenados[i]
        #si la distancia que se encuentra es menor, se actualiza la distancia minima y el par de puntos asociados
        if dist<dist_minima:
            dist_minima=dist
            mejores_puntos=(ordenados[i], ordenados[i+1])

    return (mejores_puntos,dist_minima)
```

```python
import random

LISTA_1D = random.sample(range(1, 10000), 1000)

puntos_cercanos, distancia_puntos_cercanos = dos_puntos_cercanos_divide_venceras_1d(LISTA_1D)
print(f'La distancia entre los puntos más cercanos es {distancia_puntos_cercanos} y los puntos más cercanos son: {puntos_cercanos}')
```

```python
#Se ha modificado la funcion merge_sort anterior para que se use para dos dimensiones. Se ha añadido la variable 'valor',
#que se utiliza para indicar el componente se va a ordenar.
def merge_sort(lista, valor):
    result=[]
    #si la lista solo tiene un elemento devuelve la lista
    if len(lista)<=1:
        return lista

    #Separa la lista por la mitad, la parte de la izquierda y de la derecha.
    izq_lista=merge_sort(lista[:(len(lista)//2)],valor)
    der_lista=merge_sort(lista[(len(lista)//2):],valor)

    i=0
    j=0
    #se recorre toda la lista
    for _ in range(len(der_lista)+len(izq_lista)):
        #si ya se ha gastado toda la lista derecha, se añade lo que falte a la izquierda
        if j==len(der_lista):
            result.append(izq_lista[i])
            i=i+1
        #si ya se ha gastado toda la lista izquierda, se añade lo que falte a la derecha
        elif i==len(izq_lista):
            result.append(der_lista[j])
            j=j+1
        #si los dos índices son válidos, entonces se compara las dos lista y se añade al que menos elementos tiene.
        elif der_lista[j][valor]<= izq_lista[i][valor]:
            result.append(der_lista[j])
            j=j+1
        else:
            result.append(izq_lista[i])
            i=i+1

    return result
```

```python
#Problema resuleto mediante el algoritmo de divide y vencerás.
import math
from math import inf
def dos_puntos_cercanos_divide_venceras(puntos):

    # Si el conjunto tiene 0 o 1 elementos, devolver infinito.
    if len(puntos)<=1:
        return (None, inf)
    # Si el tamaño del conjunto es 2, devolver la distancia entre ellos.
    if len(puntos)==2:
        return ((puntos[0], puntos[1]),math.dist(puntos[0], puntos[1]))

    # Ordenar los puntos
    puntos_ordenados=merge_sort(puntos,0)

    # Dividir el conjunto de puntos en dos partes iguales, la izquierda y la derecha.
    izq_conjunto_datos=puntos_ordenados[:(len(puntos)//2)]
    der_conjunto_datos=puntos_ordenados[(len(puntos)//2):]
    recta_divisora=der_conjunto_datos[0][0]
    # Solucionar el problema de forma recursiva en las partes izquierdas y derecha.
    punt_izq, dist_izq = dos_puntos_cercanos_divide_venceras(izq_conjunto_datos)
    punt_der, dist_der  = dos_puntos_cercanos_divide_venceras(der_conjunto_datos)
    if dist_izq<=dist_der:
        mejores_puntos, dist_minima=punt_izq, dist_izq
    else:
        mejores_puntos, dist_minima=punt_der, dist_der
    #Encontrar los puntos de las dos partes que se encuentran a una distancia más pequeña que la distancia minima que divide las dos partes.
    dist_horizontal=[]
    for i in puntos_ordenados:
        if abs(i[0]-recta_divisora) < dist_minima :
            dist_horizontal.append(i)
    # Encontrar la distancia mínima entre todos los pares.
    dist_vertical=merge_sort(dist_horizontal,1)
    dist_min_pares_cruzados=inf
    j=0
    for i in range(len(dist_vertical)):
        for j in range(i+1,len(dist_vertical)):
            if(dist_vertical[j][1] - dist_vertical[i][1]) >= dist_minima:
                break
            if (math.dist(dist_vertical[i],dist_vertical[j]))<dist_minima:
                    dist_minima=math.dist(dist_vertical[i],dist_vertical[j])
                    mejores_puntos= dist_vertical[i],dist_vertical[j]

    return (mejores_puntos, dist_minima)
```

```python
import random
LISTA_1D=[(random.randrange(1,10000)) for x in range(1000)]
LISTA_2D=[(random.randrange(1,10000),random.randrange(1,10000)) for x in range(1000)]
puntos_cercanos, distancia_puntos_cercanos = dos_puntos_cercanos_divide_venceras(LISTA_2D)
print(f'La distancia entre los puntos más cercanos es {distancia_puntos_cercanos} y los puntos más cercanos son: {puntos_cercanos}')
```

En este caso, utilizando este algoritmo de divide y vencerás, la complejidad es O((n^2)*log(n)). Esta complejidad si que se puede mejorar ya que si se dejan las listas ordenas durante la recursión se puede conseguir la complejidad de O(n*logn), que es la óptima para dos dimensiones.
