La función objetivo del problema consiste en maximizar la audiencia total generada en una jornada, teniendo en cuenta tanto el atractivo intrínseco de cada partido como el efecto del horario asignado y las penalizaciones derivadas de coincidencias con otros encuentros.


Así pues, la función sería:

f(x) = ∑ᵢ bᵢ · c_{h(i)} · p(k_{h(i)})

donde:

bᵢ es la audiencia base del partido i

c_{h(i)} es el coeficiente del horario asignado al partido i

k_{h(i)} es el número de partidos programados en ese mismo horario

p(k) es el factor de penalización asociado a que haya k partidos simultáneos (según la tabla dada en el problema)

Por ende, se trata un problema de maximización ya que la función objetivo es maximizar la audiencia durante la semana. Para ello, hay que ubicar los partidos en distintas zonas horarias y de esta manera se pueda distribuir la audiencia.

**Diseña un algoritmo para resolver el problema por fuerza bruta**

Para resolverlo por fuerza bruta, la idea es recorrer todas las asignaciones posibles, filtrando las que no tengan V20 y L20 (restricción), calculando la audiencia total y guardando la mejor.

```
import itertools, random

#datos
SLOTS = ["V20","S12","S16","S18","S20","D12","D16","D18","D20","L20"]
COEF  = {"V20":0.4,"L20":0.4,"S12":0.55,"S16":0.7,"S18":0.8,"S20":1.0,
         "D12":0.45,"D16":0.75,"D18":0.85,"D20":1.0}

#penalización según num de coincidencias (c = k-1). Devuelve multiplicador.
PEnALIZACIOn_COInCIDEnCIAS = {0:1.0, 1:0.75, 2:0.55, 3:0.40, 4:0.30, 5:0.25, 6:0.22, 7:0.20, 8:0.20}
def pen_from_k(k):  # k partidos en el slot
    if k <= 0: return 0.0
    c = k - 1
    return PEnALIZACIOn_COInCIDEnCIAS.get(c, 0.20)

#bases espectadores (en millones) por categorías
BASE_PAIR = {
    ("A","A"):2.0, ("A","B"):1.3, ("B","A"):1.3,
    ("A","C"):1.0, ("C","A"):1.0,
    ("B","B"):0.9, ("B","C"):0.75, ("C","B"):0.75,
    ("C","C"):0.47
}

#selection sort de índices
def selection_order_desc(valores):
    indice = list(range(len(valores)))
    for i in range(len(indice)-1):
        mejor = i
        for j in range(i+1, len(indice)):
            if valores[indice[j]] > valores[indice[mejor]]:
                mejor = j
        indice[i], indice[mejor] = indice[mejor], indice[i]
    return indice

#evaluar una asignación usando la fórmula por slots
def asignacion_puntuacion(assign, bases):
    suma_pesos = [0.0]*10
    contador  = [0]*10
    #generar las asignaciones posibles
    for i, s in enumerate(assign):
        w = bases[i]*COEF[SLOTS[s]]
        suma_pesos[s] += w
        contador[s]  += 1
    #se aplica la penalizacion del solapamiento y se suma todo.
    total = 0.0
    for s in range(10):
        total += pen_from_k(contador[s]) * suma_pesos[s]
    return total

#Fuerza bruta alg.
def brute_force(bases):
    n = len(bases)
    mejor_valor, mejor_a = -1.0, None
    for a in itertools.product(range(10), repeat=n):
        if 0 not in a or 9 not in a:  # es la restriccion, al menos debe de haber un partido en 0 y en 9
            continue
        val = asignacion_puntuacion(a, bases) #se evalua la condicion
        if val > mejor_valor: #si encuentra un valor mejor, se actualiza.
            mejor_valor, mejor_a = val, a
    return mejor_valor, mejor_a
```

**Calcula la complejidad del algoritmo por fuerza bruta**

Respuesta

La complejidad temporal del algoritmo de fuerza bruta es O(n*10^n), es decir, O(10^n), una complejidad exponencial. Por eso el problema se vuelve intratable cuando aumenta el número de partidos.

(*)**Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta**

Respuesta

Vamos a implementar un algoritmo de ramificación y poda, con el objetivo de mejorar al algoritmo de fuerza bruta.

Mientras que la fuerza bruta recorre sistemáticamente todas las posibles asignaciones, la ramificación y poda explora el mismo espacio de soluciones potenciales, pero lo hace de manera estructurada, construyendo soluciones parciales y evaluando en cada paso si merece la pena seguir profundizando en esa rama. La mejora se produce, por tanto, porque en muchos casos grandes partes del árbol de búsqueda quedan descartadas en niveles superiores, sin necesidad de llegar a las hojas.

```
#ramificacion y poda
def branch_and_bound(bases):
    n = len(bases)

    #exploramos primero los partidos "más importantes"
    order = selection_order_desc(bases) #obtener los indices de mayor a menor
    valores_base_ordenados = [bases[i] for i in order] #reordenar las bases

    #cota rápida (optimista): a lo que falte le ponemos coef 1 y sin penalización => +sum(bases_restantes)
    cota_restante = [0.0]*(n+1)
    for i in range(n-1, -1, -1):
        cota_restante[i] = cota_restante[i+1] + valores_base_ordenados[i] #cota optimista, si quedan partidos libres, aportan valor a su base.

    #estado incremental por slots
    contador  = [0]*10
    suma_pesos = [0.0]*10
    asignacion_ordenada = [0]*n
    total_actual = 0.0

    mejor_valor, mejor_asignacion = -1.0, None
    nodos = 0

    def agregar_a_slot(s, w):
        # actualiza total con delta por cambio de pen(k) y suma_pesos
        nonlocal total_actual
        antigua_k, antigua_suma = contador[s], suma_pesos[s]
        antigua_contribucion = pen_from_k(antigua_k) * antigua_suma #antigua contribucion a slot
        nueva_k, nueva_suma = antigua_k + 1, antigua_suma + w #nuevo estado tras quitar el partido
        nueva_contribucion = pen_from_k(nueva_k) * nueva_suma #nueva contribucion a slot
        contador[s], suma_pesos[s] = nueva_k, nueva_suma #actualiza el estado de slot
        total_actual += (nueva_contribucion - antigua_contribucion)

    def quitar_de_slot(s, w):
        nonlocal total_actual
        antigua_k, antigua_suma = contador[s], suma_pesos[s]
        antigua_contribucion = pen_from_k(antigua_k) * antigua_suma #antigua contribucion a slot
        nueva_k, nueva_suma = antigua_k - 1, antigua_suma - w #nuevo estado tras quitar el partido
        nueva_contribucion = pen_from_k(nueva_k) * nueva_suma #nueva contribucion a slot
        contador[s], suma_pesos[s] = nueva_k, nueva_suma #actualiza el estado de slot
        total_actual += (nueva_contribucion - antigua_contribucion)

    def dfs_busqueda(pos, usa_slot_inicial, usa_slot_final):
        nonlocal mejor_valor, mejor_asignacion, nodos
        nodos += 1

        #poda por cota superior optimista
        if total_actual + cota_restante[pos] <= mejor_valor:
            return

        if pos == n:
            if usa_slot_inicial and usa_slot_final and total_actual > mejor_valor:
                #reconstruir en orden original
                a = [0]*n
                for p, indice in enumerate(order):
                    a[indice] = asignacion_ordenada[p]
                mejor_valor, mejor_asignacion = total_actual, tuple(a)
            return

        restantes = n - (pos + 1)
        base = valores_base_ordenados[pos]

        for s in range(10):
            nuevo_slot_0 = usa_slot_inicial or (s == 0)
            nuevo_slot_9 = usa_slot_final or (s == 9)

            #si ya no quedan partidos después, forzar que se cumpla
            if restantes == 0 and (not nuevo_slot_0 or not nuevo_slot_9):
                continue

            w = base * COEF[SLOTS[s]]
            asignacion_ordenada[pos] = s
            agregar_a_slot(s, w)
            dfs_busqueda(pos+1, nuevo_slot_0, nuevo_slot_9)
            quitar_de_slot(s, w)

    dfs_busqueda(0, False, False)
    return mejor_valor, mejor_asignacion, nodos
```

(*)**Calcula la complejidad del algoritmo**

Respuesta

En cuanto a la complejidad del algoritmo de ramificación y poda, en el peor caso sigue siendo exponencial. Si tenemos n partidos y m horarios, el árbol completo tiene mⁿ hojas, exactamente igual que en fuerza bruta. Por tanto, la complejidad teórica en el peor caso es O(mⁿ).

Sin embargo, la diferencia fundamental es que, en la práctica, muchas ramas del árbol no se exploran gracias a la poda. Si en un nodo parcial la mejor audiencia que podría alcanzarse (según una cota superior optimista) es menor o igual que la mejor solución completa ya encontrada, esa rama se descarta sin generar todas sus extensiones. Por ello, aunque la cota asintótica no mejora en el peor caso, el número real de nodos explorados suele ser (y será como veremos en apartados posteriores),muchísimo menor que mⁿ.

**Según el problema (y tenga sentido), diseña un juego de datos de entrada aleatorios**

Respuesta

```
def random_bases(n=10, seed=0):
  random.seed(seed)
  cats = ["A"]*3 + ["B"]*11 + ["C"]*6
  bases = []
  for _ in range(n):
    c1 = random.choice(cats)
    c2 = random.choice(cats)
    bases.append(BASE_PAIR[(c1,c2)])
  return bases
```

En primer lugar, la función recibe como parámetro el número de partidos n, lo que permite analizar el comportamiento del algoritmo a distintas escalas del problema.

La utilización de un parámetro seed para inicializar el generador de números aleatorios garantiza la reproducibilidad de los resultados. Sin esta semilla, cada ejecución produciría una instancia distinta y la comparación perdería rigor.

El vector cats está diseñado para reflejar una distribución realista de categorías de equipos. En lugar de asignar las categorías con la misma probabilidad, se construye una lista ponderada donde la categoría A aparece 3 veces, la B 11 veces y la C 6 veces. Esto implica que, al seleccionar aleatoriamente mediante random.choice, la probabilidad de que un equipo pertenezca a cada categoría es proporcional a su frecuencia en la lista.




Para cada partido se seleccionan dos categorías de forma independiente y aleatoria, generando así combinaciones diversas (A-A, A-B, B-C, etc.). Posteriormente, la audiencia base del enfrentamiento se obtiene a través del diccionario BASE_PAIR, que traduce cada pareja de categorías en una estimación numérica de audiencia.

El resultado final es una lista bases de tamaño n que contiene exclusivamente las audiencias base de los partidos generados.

Estos resultados permiten justificar que, aunque la complejidad teórica en el peor caso sigue siendo exponencial, la complejidad real en la práctica se reduce de manera drástica gracias a la poda.

En resumen, los resultados confirman que la ramificación y poda mejora de forma muy notable el rendimiento respecto a fuerza bruta, especialmente a medida que crece el tamaño del problema, manteniendo al mismo tiempo la garantía de encontrar el óptimo global.
